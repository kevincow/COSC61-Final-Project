```python
import pandas as pd
import numpy as np
import sklearn as sk
from sklearn import cluster, metrics
from sklearn.cluster import KMeans
from scipy.stats import linregress
import matplotlib.pyplot as plt
```

```python
# Question 4
df_strikeouts_time = pd.read_csv("strikeouts_over_time.csv")
df_strikeouts_time = df_strikeouts_time.sort_values(by='season')
```

```python
# Question 2
df_attendance_full = pd.read_csv("attendance_vs_length.csv")
df_attendance_30 = pd.read_csv("attendance_vs_length_30yr.csv")
```

```python
# Question 2

# Create a line plot using Matplotlib
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
plt.scatter(df_attendance_30['length_minutes'], df_attendance_30['attendance'], marker='o', color='b', label='Strikeouts')

# Calculate OLS regression line
slope, intercept, r_value, p_value, std_err = linregress(df_attendance_30['length_minutes'], df_attendance_full['attendance'])
regression_line = [slope * x + intercept for x in df_attendance_30['length_minutes']]

plt.plot(df_attendance_30['length_minutes'], regression_line, color='r', label='OLS Regression Line')

# Add labels and title
plt.xlabel('Length (Minutes)')
plt.ylabel('Attendance')
plt.title('Attendance vs. Game Length (Last 30 Years)')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Question 2

# Create a line plot using Matplotlib
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
plt.scatter(df_attendance_full['length_minutes'], df_attendance_full['attendance'], marker='o', color='b', label='Strikeouts')

# Calculate OLS regression line
slope, intercept, r_value, p_value, std_err = linregress(df_attendance_full['length_minutes'], df_attendance_full['attendance'])
regression_line = [slope * x + intercept for x in df_attendance_full['length_minutes']]

plt.plot(df_attendance_full['length_minutes'], regression_line, color='r', label='OLS Regression Line')

# Add labels and title
plt.xlabel('Length (Minutes)')
plt.ylabel('Attendance')
plt.title('Attendance vs. Game Length (All Games)')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
